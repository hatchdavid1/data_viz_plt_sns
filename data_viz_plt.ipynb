```
plt.figure(figsize=(10,5)) # Tamaño de la gráfica 
x = np.array([1,2,3,4,5,6]) # Generando un array del 1 al 6
y1 = np.power(x,2) # función praa elevar x**2
y2 = np.power(x,3) # Función para elevar x**3
plt.plot(x,y1,'b-', label = '$y1 = x^2$') # Nombre de la función y color b- azul
plt.plot(x,y2, 'r-', label = 'y2 = x^3$') # Nombre de la función y color r- rojo
plt.xlabel('X - Axis') # Título del eje x
plt.ylabel('Y - Axis') # Título del eje y 
plt.legend() # Agregar nombres de las funciones 
plt.tight_layout() # Ajusta el tamaño entre funciones
plt.show() # Mostrar el gráfico 
```

```
x = np.linspace(0,10,2000) # generando 2000 números entre 0 y 10
plt.figure(figsize=(10,6)) # Tamaño del gráfico 
plt.plot(x, np.sin(x), label = '$Sin(X)$') # Gráfico del seno de x con nombre
plt.plot(x, np.cos(x), label = '$Cos(X)$') # Gráfico del coseno de x con nombre
plt.xlabel(r'$X$', fontsize = 18) # Formato para título del eje x
plt.ylabel(r'$Y$', fontsize = 18) # Formato para títutlo del eje y 
plt.title('$Sin(X) $ $ & $ $ Cos(X)$', fontsize = 14) # Foramto para título de la gráfica
plt.legend(loc = 'upper right' ) # Posición de referencia de funciones 
plt.show() # Mostrar gráfico
```

```
# changing the line style
plt.figure(figsize=(10,5)) # Tamaño de la gráfica 
x = np.array([1,2,3,4,5,6]) # Generar un array del 1 al 6 
y1 = np.power(x,2) # Función para elevar x**2
y2 =  np.power(x,3) # función para elevar x**3
plt.plot(x,y1, 'b-', label = '$y1 = x^2$') # Títulos de línea con color b- azul
plt.plot(x, y2, color = 'red', linewidth= 1.0, linestyle= '--') # Títulos de línea con color red y estilo --
plt.xlabel('X - Axis') # Título del eje x 
plt.ylabel('Y- Axis') #Título del eje y 
plt.legend(loc = 'upper center', fontsize = 'large') # Nombres de funciones con posición central 
plt.show() #Mostrar gráfica 
```

```
# Print de múltiples gráficas en una figura
plt.figure(figsize=(12,6))
x = np.linspace(0, 10, 100) 
y1 = np.sin(x) # Gráfica Seno
y2 = np.cos(x) # Grafic Coseno

plt.subplot(2,1,1)
plt.plot(x,y1, "b-")

plt.subplot(2,1,2)
plt.plot(x,y2, "r-")
plt.tight_layout()
plt.show()
```
